```python
from pymongo import MongoClient
import pprint

import pandas as pd

import numpy as np

# Requests sends and recieves HTTP requests.
import requests

# Beautiful Soup parses HTML documents in python.
from bs4 import BeautifulSoup

import json
import time
```

```python
client = MongoClient('localhost', 27017)
db_aud = client['aud_scraper']

rates = db_aud['rates']
```

```python
url = 'https://www.rba.gov.au/statistics/cash-rate/'
```

```python
r = requests.get(url)
r.status_code
```

```python
pprint.pprint(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
links = soup.find_all('tr')
```

```python
def scrape_AUD(links):
    
    """
    PARAMETERS:
    links - a list of parsable html tags generated by BeautifulSoup, pertinent to the Reserve Bank of Australia's website
    
    RETURNS:
    None - uploads rows of rate data to MongoDB
    """

    for link in links[1:]:
        d= dict()

        # compile dates
        for idx, i in enumerate(link.find_all('th')):
            d['date'] = i.text

        # compile rate change and target rate
        for idx, i in enumerate(link.find_all('td')):
            if idx == 0:
                d['change%'] = i.text

            elif idx == 1:
                d['AUD_rate_tgt%'] = i.text


            else:
                continue

        rates.insert_one(d)

        time.sleep(1)
    
    return None
    
```

```python
scrape_AUD(links)
```
